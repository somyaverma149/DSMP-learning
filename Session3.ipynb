```python
# scalar multiplication on a vector
v = [2,3,4]
s = [-3]

x = []
for i in v:
  x.append(i*s)
print(x)
```

```python
L = [s*i for i in v]
print(L)
```

```python
L = [1,2,3,4,5]
L = [i**2 for i in L]
print(L)
```

```python
L = [i for i in range(1,51) if i%5 == 0]
print(L)
```

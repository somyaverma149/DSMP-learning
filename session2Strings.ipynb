```python
n = int(input('enter the num :'))
fact = 1
for i in range(1,n+1):
  fact = fact*i
print(fact)
```

```python
# Sequence Sum

# 1/1! + 2/2! + 3/3! + .........
import math as m

n = int(input('enter n'))
x = 1
for i in range(2,n+1):
  x = x+(i/m.factorial(i))
print(x)

```

Nested Lops

```python
for i in range(1,6):
  for j in range(1,6):
    print(i,j)
```

```python
rows = int(input('enter the num of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end ='')
  print()
```

```python
rows = int(input('enter the rows'))

for i in range(1,rows + 1):
  for j in range(1,i+1):
    for k in range(1,j+1):
      print(k,end ='')
    print()
```

Break

```python
# Break
for i in range(1,10):
  if i ==6:
    break
  print(i)
```

```python
lower = int(input('enter lower range'))
upper = int(input('enter upper range'))

for i in range(lower,upper + 1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)
```

```python
# Continue Statement
for i in range(1,10):
  if i ==5:
    continue
  print(i)
```

```python
for i in range(1,10):
  pass
```

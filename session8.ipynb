Pandas series

```python
import numpy as np
import pandas as pd
```

```python
Pandas is a fast,powerful , flexible and easy to use  open source data analysis and manipulation tool,built on top of the python programming language.
It is a 1d array.
```

```python
# strings
country = ['India','Nepal','USA','Pakistan','Srilanka']


pd.Series(country)
```

```python
# integers
runs = [1,2,3,4,5,6]


pd.Series(runs)
```

```python
marks = [67,58,89,99]
subjects = ['maths','english','hindi','science']

pd.Series(marks,index=subjects)
```

```python
# setting a name
pd.Series(marks,index=subjects,name='nitish k marks')
```

```python
marks = pd.Series(marks,index=subjects,name='nitish k marks')

marks
```

```python
marks = {
  'maths' : 67,
  'english' : 58,
  'hindi' : 89,
  'science' : 98
}

marks_series = pd.Series(marks,name='nitish k marks')
marks_series
```

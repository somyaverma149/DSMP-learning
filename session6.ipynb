Stacking

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

a4
a5
```

```python
np.hstack((a4,a5))
```

Splitting

```python
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

a4
a5
```

```python
# horizontal splitting
np.hsplit(a4,2)
```

```python
# vertical splitting
np.vsplit(a4,3)
```
